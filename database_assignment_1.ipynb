2. Explain the purpose of constraints and how they help maintain data integrity in a database. Provide
examples of common types of constraints.


ans:-Purpose of Constraints

Constraints in a database are rules that are applied to the data to ensure its accuracy, consistency, and reliability. They help maintain data integrity by preventing invalid data from being entered or updated in the database. By enforcing these rules, constraints ensure that the data stored in the database is always valid and meaningful.

How Constraints Help Maintain Data Integrity

Constraints act as safeguards against data corruption and inconsistencies. They ensure that the data conforms to predefined business rules and requirements. This is crucial for maintaining the quality and trustworthiness of the data stored in the database.

Examples of Common Types of Constraints

Here are some common types of constraints used in databases:

NOT NULL: This constraint ensures that a column cannot contain NULL values. It enforces the presence of data in a specific column.

Example: emp_name TEXT NOT NULL in the employees table would prevent inserting a record without an employee name.

UNIQUE: This constraint ensures that all values in a column or group of columns are unique. It prevents duplicate entries in the specified column(s).

Example: email TEXT UNIQUE in the employees table would ensure that each employee has a unique email address.

PRIMARY KEY: This constraint uniquely identifies each record in a table. It combines the properties of NOT NULL and UNIQUE constraints. A table can have only one primary key.

Example: emp_id INT PRIMARY KEY NOT NULL in the employees table designates emp_id as the primary key, ensuring each employee has a unique identifier.

FOREIGN KEY: This constraint establishes a relationship between two tables by linking a column in one table (the child table) to the primary key column in another table (the parent table). It ensures referential integrity by preventing invalid relationships between tables.

Example: If you had a departments table with a primary key dept_id, you could add a dept_id column to the employees table and define it as a foreign key referencing the departments table. This would ensure that each employee is associated with a valid department.

CHECK: This constraint allows you to define a specific condition that must be met for data to be inserted or updated in a column. It provides flexibility in defining custom validation rules.

Example: age INT CHECK (age >= 18) in the employees table ensures that all employees are at least 18 years old.

DEFAULT: This constraint provides a default value for a column when no value is specified during insertion. It ensures that a column always has a value, even if it's not explicitly provided.

Example: salary DECIMAL DEFAULT 30000 in the employees table sets a default salary of 30,000 for new employees if their salary is not specified.

Data Consistency: By enforcing the presence of a value, it helps maintain data consistency across the database.

Query Optimization: Queries that rely on the column can be optimized, as the database can assume that the column always contains a value.
Can a Primary Key Contain NULL Values?

No, a primary key cannot contain NULL values. This is a fundamental principle of relational database design. Here's why:

Unique Identification: The primary key is used to uniquely identify each record in a table. If a primary key could be NULL, it would be impossible to distinguish between records with NULL values in the primary key column.
Referential Integrity: Primary keys are often used as foreign keys in other tables to establish relationships. Allowing NULL values in a primary key would break referential integrity, as foreign keys would not be able to reliably link to the primary key.
Data Consistency: By ensuring that primary keys are always unique and non-null, it helps maintain data consistency and prevents data corruption.
Justification

The requirement for primary keys to be non-null is enforced by the database management system (DBMS). When you define a primary key constraint, the DBMS automatically adds a NOT NULL constraint to the primary key column(s). This ensures that the fundamental principles of primary keys are always upheld.

Therefore, applying the NOT NULL constraint to a column, especially a primary key column, is crucial for maintaining data integrity, consistency, and referential integrity in a database.

5. Explain the consequences of attempting to insert, update, or delete data in a way that violates constraints.
Provide an example of an error message that might occur when violating a constraint.

ans:-Consequences of Violating Constraints

When you attempt to insert, update, or delete data in a way that violates a constraint, the database management system (DBMS) will reject the operation to protect the integrity of the data. This means the data will not be modified, and you will receive an error message indicating the violation.

Here's a breakdown of the consequences for each type of operation:

Insert: If you try to insert a row with a duplicate value for a unique constraint or a primary key, the insertion will be rejected. Similarly, if you try to insert a row with a NULL value in a column with a NOT NULL constraint, the insertion will fail.
Update: If you attempt to update a row in a way that violates a constraint (e.g., changing a value to a duplicate or setting a required column to NULL), the update will be rejected.
Delete: If you try to delete a row that is referenced by a foreign key in another table (violating referential integrity), the deletion will be blocked.
Example Error Message

The specific error message you receive will depend on the DBMS you're using and the type of constraint violated. However, here's a general example of an error message you might encounter when violating a unique constraint:


ERROR:  duplicate key value violates unique constraint "UQ_email"
DETAIL:  Key (email)=(john.doe@example.com) already exists.
Use code with caution
In this example, the error message clearly indicates that the operation failed because it would have created a duplicate value in the email column, which is protected by a unique constraint named UQ_email.
