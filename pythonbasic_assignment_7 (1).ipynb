```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([3, 7, 9, 15, 22, 29, 35])

x = np.arange(len(data))  # Create x-axis values (0, 1, 2, ...)

# Scatter plot
plt.scatter(x, data)

# Trend line (linear regression)
coefficients = np.polyfit(x, data, 1)  # Fit a 1st-degree polynomial (line)
trendline = np.poly1d(coefficients)  # Create a function for the trend line
plt.plot(x, trendline(x), color='red', linestyle='--')

plt.xlabel("Index")
plt.ylabel("Data Value")
plt.title("Scatter Plot with Trend Line")
plt.show()
```

5.generate a synthetic dataset with correlated features. visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate correlated data
np.random.seed(42)  # For reproducibility
num_samples = 100
num_features = 5

# Create a random correlation matrix
correlation_matrix = np.random.rand(num_features, num_features)
correlation_matrix = (correlation_matrix + correlation_matrix.T) / 2  # Make it symmetric
np.fill_diagonal(correlation_matrix, 1)  # Set diagonal to 1

# Generate data using Cholesky decomposition
cholesky_decomposition = np.linalg.cholesky(correlation_matrix)
data = np.random.randn(num_samples, num_features) @ cholesky_decomposition

# Create a pandas DataFrame
df = pd.DataFrame(data, columns=[f"Feature {i}" for i in range(1, num_features + 1)])

# Calculate correlation matrix
correlation_matrix = df.corr()

# Visualize correlation matrix using heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

2.create a bokeh scatter plot using randomly generated x and y values .use different sizes and colors for the markers based on the sizes and colors coulmns .

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Generate random data
np.random.seed(42)  # For reproducibility
num_points = 50
data = {
    'x': np.random.rand(num_points),
    'y': np.random.rand(num_points),
    'sizes': np.random.randint(5, 20, size=num_points),  # Random sizes between 5 and 20
    'colors': np.random.choice(['red', 'blue', 'green'], size=num_points)  # Random colors
}
df = pd.DataFrame(data)

# Create ColumnDataSource
source = ColumnDataSource(df)

# Create Bokeh plot
p = figure(title="Scatter Plot with Random Data", x_axis_label="x", y_axis_label="y")
p.scatter(x='x', y='y', size='sizes', color='colors', source=source)

# Show plot
show(p)
```

5.create a bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)

xx, yy = np.meshgrid(x, y)

ans:-

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar, BasicTicker, PrintfTickFormatter, ColumnDataSource

# Data (provided)
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create Bokeh plot
p = figure(title="Heatmap", x_axis_label="x", y_axis_label="y",
           x_range=(x.min(), x.max()), y_range=(y.min(), y.max()),
           toolbar_location=None, tools="")  # Remove toolbar and tools

# Reshape data for ColumnDataSource
# We need to flatten the 2D arrays into 1D and create a dictionary
data = {'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()}

# Create ColumnDataSource
source = ColumnDataSource(data)


# Add heatmap using rect glyphs
# Use the ColumnDataSource as the source
p.rect(x="x", y="y", width=0.1, height=0.1, source=source, # Changed to use source
       fill_color={'field': 'value', 'transform': color_mapper},
       line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, ticker=BasicTicker(),
                     formatter=PrintfTickFormatter(format="%0.2f"),
                     label_standoff=12, border_line_color=None, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show plot
show(p)
```
